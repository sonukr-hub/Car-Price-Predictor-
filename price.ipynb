# Cleaning Data

```python
# backup 
data=df.copy()
```

```python
data=data[data['year'].str.isnumeric()]
```

```python
data['year']=data['year'].astype(int)
```

```python
data.info()
```

```python
data=data[data['Price']!="Ask For Price"]
```

```python
data['Price']=data['Price'].str.replace(',','').astype(int)
```

```python
data.info()
```

```python
data['kms_driven']=data['kms_driven'].str.replace(',','')
data['kms_driven']=data['kms_driven'].str.replace(' kms','')
```

```python
data=data[data['kms_driven'].str.isnumeric()]
data['kms_driven']=data['kms_driven'].astype(int)
```

```python
data.info()
```

```python
data=data[~data['fuel_type'].isna()]
```

```python
data.info()
```

```python
data['name']=data['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
data=data.reset_index()
data
```

```python
print(data.info())
```

```python
data.describe()
```

```python

data=data[data['Price']<6e6].reset_index()
data
```

```python
# save Our work
data.to_csv('cleaned_car.csv')
```

# Preprocessing/Making For Linear Regression 
# Creating Model and Training

```python
X=data[['name','company','year','kms_driven','fuel_type']]
y=data['Price']
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer(
    (OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_predict=pipe.predict(X_test)
y_predict
```

```python
r2_score(y_test,y_predict)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_predict=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_predict))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```
